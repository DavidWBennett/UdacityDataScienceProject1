# Are Guests More Likely to Review If They Want To Give a Very High (5 stars) or Very Low (1 Star) Review?

### Import Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

FilePathRoot = "C:\\Users\\USBennetDa2\\OneDrive - NESTLE\\Udacity Data Science Nanodegree\\Project 1"
```

### Analysis

```python
# Listing Data for Number of Reviews and the Average Score Review
ListingsCleaned = pd.read_csv(FilePathRoot + "\\ListingsCleaned.csv")

# Is there a correlation between the number of review and the review rating for Seattle AirBnBs?
print(ListingsCleaned[['number_of_reviews', 'review_scores_rating']].corr())
```

### Visuals

```python
# Create a scatter plot.
plt.scatter(ListingsCleaned[["number_of_reviews"]], ListingsCleaned[["review_scores_rating"]])

# Add in additional characteristics of the graph, such as the legend and title.
plt.ylabel('Average Review Score (Out of 100)');
plt.xlabel('Number of Reviews');
plt.title("Are Guests More Likely To Review If they Had a Very Positive or Very Negative Experience?");
```
